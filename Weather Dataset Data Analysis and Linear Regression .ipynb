# Division into training and testing dataset

```python
train = df[df['DATE']<20000125].copy()
```

```python
train
```

```python
test = df[df['DATE']>=20000125].copy()
```

```python
test
```

# Applying Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
independent_var = ["BASEL_global_radiation","BASEL_sunshine"]
dependent_var = "BASEL_cloud_cover"
```

```python
reg.fit(train[independent_var], train['BASEL_cloud_cover'])
```

```python
predictions = reg.predict(test[independent_var])
```

```python
predictions
```

```python
test["predictions"] = predictions
```

```python
test
```

```python
test['predictions'] = test['predictions'].round()
```

```python
test
```

# Calculating error in prediction

```python
from sklearn.metrics import mean_absolute_error
```

```python
error = mean_absolute_error(test['BASEL_cloud_cover'],test['predictions'])
```
